```python
#lam sach du lieu ngay gio
def proc_s(x):
  s1 = str(x).split(" ")
  if len(s1) == 2:
    return s1[0]
  else:
    return x; #chu y

df1['join date'] = df1['join date'].apply(proc_s) # tu dong chay



df1
```

```python
df1
```

```python

# for idx,x in enumerate(df1['join date']):
#     if str(x).startswith('[') and str(x).endswith(']'):
#       df1['join date'] = df1['join date'].astype().replace('[',' ').replace(']',' ')

df1['join date'] = df1['join date'].astype(str).apply(lambda x: x[1:len(x) - 1] if x.startswith('[') and x.endswith(']') else x)
df1
```

```python
def xuLyNgay(x):
  s1 = str(x).split('-')
  if len(s1[0]) == 2:
    s2 = s1[2] + '-' + s1[1] + '-' + s1[0]
    return s2
  else:
    s3 = str(x).split(" ")
    return s3[0]

df1['join date'] = df['join date'].apply(xuLyNgay)
df1
```

```python
def proc_date(x):
  s1 = str(x).split("-")
  day = int(s1[2])
  month = int(s1[1])
  if 1 <= day <= 31 and 1 <= month <= 12:
    return True
  else:
    return False

k = []
for idx, x in enumerate(df1['join date']):
  if proc_date(x) == False:
    k.append(idx)

df1 = df1.drop(k)
df1
```

```python
#Xu ly du lieu ten
def proc_str(x):
  s1 = str(x).split(" ")
  strb = ""
  for k in s1:
    s2 = k[0].upper() + k[1:]
    strb += s2 + " "
  return strb




# df1['full name'] = df1['full name'].apply(proc_s)
# df1



df1['full name'] = df1['full name'].astype(str).apply(lambda x: x.lower())
df1['full name'] = df1['full name'].astype(str).apply(proc_str)
df1
# df1
```

```python
df1.reset_index(drop = True, inplace = True)
df1
```

```python
#Tao them 1 cot email voi format '{last_name}.{frist_name}.{id}@gmail.com'
email_list = []
for i in range(len(df1)):
    str1 = df1['full name'][i].split(" ")
    lastName = str1[0]
    fristName = str1[1]
    id1 = str(df['cust id'][i])
    email = '{last_name}.{frist_name}.{id}@gmail.com'.format(last_name = lastName, frist_name = fristName, id = id1)
    email_list.append(email)

df1['email'] = email_list

df1
```

```python
# Format the p(hone follow template {84.......}
def proc_phone(x):
  return "84" + str(x)

df1['phone'] = df1['phone'].astype(str).apply(lambda x: x if x.startswith("84") else proc_phone(x))
df1
```
